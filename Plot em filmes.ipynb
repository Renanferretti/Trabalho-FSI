```
import pandas as pd
```

```
import seaborn as sns
```

```
filmes = pd.read_csv('movies.csv', encoding = 'latin-1')
```

```
#info sobre as colunas
filmes.info()
```

```
#qual gênero predominante entre filmes considerados muito bons ? 
sns.countplot(x = "genre", data = filmes[filmes.score>8.5])
```

```
#verificando quais gêneros são os mais vistos nos últimos anos
sns.countplot(y = "genre", data = filmes[filmes.year>2000])
```

```
''' vendo a plotagem anterior e comparando com essa, percebemos uma 
soberania de décadas '''
sns.countplot(y = "genre", data = filmes[filmes.year<2000])
```

Os gêneros comédia, ação e drama predominam, contendo filmes como: O Poderoso Chefão, Coringa, John Wick, Se beber não case

```
#nota dos filmes com arrecadações milionárias - 400 milhões
sns.countplot(x="score", data = filmes[filmes.gross>400000000])
```

```
filmes.score.median()
```

```
#USA abriga a grande maioria dos filmes bem avaliados pela crítica
sns.countplot(y='country', data=filmes[filmes.score>8.5])
```

```
#USA tem filmes bons porque tem bons atores
#Elijah Wood é interpreta Frodo em Senhor dos Anéis
sns.countplot(y='star', data=filmes[filmes.score>8.5])
```

```
#ainda possuem os melhores diretores
sns.factorplot(y='director', col='country', kind='count', data=filmes[filmes.score>8.5])
```

```
# 11 milhões
filmes.budget.median()
```

```
#por fim possuem um orçamento bem maior que a média, 4 milhões acima
filmes[filmes.country == 'USA'].budget.median()
```

```
sns.countplot(y='company', data= filmes[filmes.score>8.5])
```

Toda a predominância dos EUA no ramo cinematográfico, acaba atraindo tanto atores quanto diretores

```
#gêneros que Nicolas Cage mais atuou, que são os três gêneros com mais filmes
sns.countplot(y="genre", data = filmes[filmes.star == 'Nicolas Cage'])
```

```
#ano a ano vendo quais gêneros lucraram mais
sns.factorplot(y='genre', col='year', kind='count', data=filmes[filmes.gross>filmes.gross.median()])
```

Nos últimos anos, o gênero ação possui mais filmes com lucro acima da média do que os filmes de comédia, que dominaram durante décadas esse quesito

```
#filmes com nota acima da média possuem entre 100 e 150 min
filmes[filmes.score>filmes.score.median()].plot.scatter(x='score', y ='runtime')
```

Quanto mais perto dos 100 min, maior é a avaliação. São poucos casos que obtiveram nota acima da média estando longe dos 100 min

filmes de ação,comédia e drama feitos nos EUA, que possuem a melhor estrutura 
possível para se fazer um filme, são os mais vistos. A melhores avaliações são dadas para filmes perto dos 100 min
